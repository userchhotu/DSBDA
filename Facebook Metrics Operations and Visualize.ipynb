# Part 2 (Slip 4) Visualize the facebook Data

```python
# Import Required Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r'C:\Users\lokes\OneDrive\Desktop\abc\Facebook.csv',delimiter=';')
```

```python
df.head()
```

```python
## Line Graph
plt.figure(figsize=(8,5))

plt.plot(df['like'].head(20))

plt.title("Likes Trend")
plt.xlabel("Posts")
plt.ylabel("Likes")
plt.show()
```

```python
# Bar graph
type_likes = df.groupby('Type')['like'].mean()

plt.figure(figsize=(8,5))

type_likes.plot(kind='bar')

plt.title("Average Likes by Post Type")
plt.xlabel("Post Type")
plt.ylabel("Average Likes")

plt.show()
```

```python
# Histogram
plt.figure(figsize=(8,5))

plt.hist(df['like'], bins=20)

plt.title("Distribution of Likes")
plt.xlabel("Likes")
plt.ylabel("Frequency")

plt.show()
```

```python
#scatter plot
plt.figure(figsize=(8,5))

plt.scatter(df['like'], df['share'])

plt.title("Likes vs Shares")
plt.xlabel("Likes")
plt.ylabel("Shares")

plt.show()
```

```python
# box plot using seaborn
plt.figure(figsize=(8,5))

sns.boxplot(x=df['like'])

plt.title("Boxplot of Likes")

plt.show()
```

```python
#  Heatmap using Seaborn
plt.figure(figsize=(12,8))

corr = df.corr(numeric_only=True)

sns.heatmap(corr, annot=True)

plt.title("Correlation Heatmap")

plt.show()
```

```python
# Pair Plot using Seaborn
sns.pairplot(df[['like', 'share', 'comment']])

plt.show()
```
